```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.xkcd()

%pylab inline
pylab.rcParams['figure.figsize'] = (20, 12)
```

```python
a = np.random.logistic(size=100)
```

```python
a[a<-2] = 0
```

```python
plt.plot(np.sort(a), linewidth=5)
plt.axis('off')


plt.annotate('LOGISTIC REGRESSION',
             xy=(0, -2), arrowprops=dict(arrowstyle='->'), xytext=(0, 3), size=20)

plt.annotate('DECISION TREE',
             xy=(20, -1), arrowprops=dict(arrowstyle='->'), xytext=(8, -2), size=20)

plt.annotate('SVM (LINEAR KERNEL)',
             xy=(25, 0), arrowprops=dict(arrowstyle='->'), xytext=(20, 1), size=20)

plt.annotate('NAIVE BAYES',
             xy=(40, -0.5), arrowprops=dict(arrowstyle='->'), xytext=(40, -1.5), size=20)

plt.annotate('TREE ENSEMBLE MODELS',
             xy=(70, 0.5), arrowprops=dict(arrowstyle='->'), xytext=(70, -0.5), size=20)

plt.annotate('NEURAL NETWORKS',
             xy=(80, 6), arrowprops=dict(arrowstyle='-'), xytext=(60, 4), size=20)


plt.show()
```
